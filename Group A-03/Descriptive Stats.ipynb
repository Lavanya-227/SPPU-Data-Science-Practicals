```python
#Import Libraries
import numpy as np
import pandas as pd
```

```python
#1. 'data-income-age-group.csv' dataset
df1 = pd.read_csv('data-income-age-group.csv')
df1.head()
```

```python
df1.isnull().sum()
```

```python
df1.columns
```

```python
# Grouping by the 'Age' column and calculating summary statistics for 'Income($)'
df1.groupby('Age')['Income($)'].describe()
```

```python
#a list that contains a numeric value for each response to the categorical variable
edu_cat = df1['Name']
edu_num_vals, edu_cats_unique = pd.factorize(edu_cat)
print("Numeric values for each unique category:")
print(dict(zip( edu_cats_unique, range(len( edu_cats_unique)))))
print("List of numeric values")
print(edu_num_vals)
```
